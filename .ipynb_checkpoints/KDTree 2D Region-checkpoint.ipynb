```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification, make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.neighbors import KDTree, BallTree, NearestNeighbors
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0,
                           n_clusters_per_class=1, n_classes=2, class_sep=2, random_state=7)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[: , 0] , X[: , 1] , c = y , cmap='jet' , ec='k' , s=100);
```

```python
kdt = KDTree(X, leaf_size=40)
```

```python
bounds = kdt.node_bounds.base
bounds = np.vstack(kdt.node_bounds.base)
nodes = kdt.node_data.base
print('bounds:', bounds)
print('nodes:', nodes)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X[: , 0] , X[: , 1] , c = y , cmap='jet' , ec='k' , s=100);

for bound in bounds[: , 0]:
    plt.axvline(bound, linewidth=2, color='r')
    
for bound in bounds[: , 1]:
    plt.axhline(bound, linewidth=2, color='b')
```
